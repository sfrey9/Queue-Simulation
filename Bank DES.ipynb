```python
# Modelling and Simulation 3911b
# Assignment # 4
# Shardul Frey
# 250696234
# March 27th 2016
#
# Summary:
#
# This program simulates the processing of customers at a bank
#
# notes: 
#
# In this model during each timestep a customer can leave the teller 
# and a new customer can approach the teller during the same time step.
#
tellerstate = False; 
# the tellerstate is true if occupied, false if unnocupied.
tellertime = 0;
# teller time will equal the total amount of time the teller is 
# occupied. This should equal the sum of all the service times.
simtime = 0;
# this will keep track of the total time the simulation has been running
tinterval = 0.1;
# the starting time is 0 and the tick rate is 0.1 minutes.

c1 = [1, 3.2, 3.8, 0, 0,0,0,0,0];
c2 = [2, 10.9, 3.5, 0, 0,0,0,0,0];
c3 = [3, 13.2, 4.2, 0, 0,0,0,0,0];
c4 = [4, 14.8, 3.1, 0, 0,0,0,0,0];
c5 = [5, 17.7, 2.4, 0, 0,0,0,0,0];
c6 = [6, 19.8, 4.3, 0, 0,0,0,0,0];
c7 = [7, 21.5, 2.7, 0, 0,0,0,0,0];
c8 = [8, 26.3, 2.1, 0, 0,0,0,0,0];
c9 = [9, 32.1, 2.5, 0, 0,0,0,0,0];
c10 = [10, 36.6, 3.4, 0, 0,0,0,0,0];

# entry 1: Customer ID
# entry 2: Arrival time
# entry 3: Service time
# entry 4: Customer state. 0 if customer hasn't arrived
#                          1 if customer is in queue
#                          2 if customer is being served
#                          3 if customer has departed
# entry 5: Customer position in queue. 0 if customer not in queue.
# entry 6: Time of commencement of service
# entry 7: Departure time
# entry 8: Time in queue
# entry 9: Time in bank

allc = [c1,c2,c3,c4,c5,c6,c7,c8,c9,c10];
# the customers for the simulation are initialized we are now ready to start.

queue = 0;
# the initial length of the queue is 0 people.

queuem = 0;
# queuem will be used to calculate the average length of the queue.
servicecount = 0;
# counts the time spent being serviced

for f in range (0,401):    
    for i in range (0,10):
        
        # This block assigns new arrivals to either the teller
        # or the end of the queue depending on whether the teller
        # is occupied and if there's a queue.
        if allc[i][1] == simtime:
            if (tellerstate == False and queue == 0):
                allc[i][3] = 2;
                allc[i][5] = simtime;
                tellerstate = True;
                #reset the service timer
                servicecount = 0
            elif (tellerstate == True):
                queue += 1;
                allc[i][4] = queue;
                allc[i][3] = 1;
                
        # This block checks if the customer at the teller is done. 
        # If so the customer is sent on their way.
        if (tellerstate == True and allc[i][3] == 2 and servicecount == allc[i][2]):
            tellerstate = False;
            allc[i][3] = 3;
            allc[i][6] = simtime;
            tellertime += servicecount;
            servicecount = 0;
            
        
    # This block checks if the teller is occupied and if he/she
    # isn't assigns the next person in the queue to the teller.
    if (tellerstate == False and queue >= 1):
        for i in range (0,10):
            if allc[i][4] == 1:
                allc[i][3] = 2;
                allc[i][5] = simtime;
                tellerstate = True;
            if allc[i][4] >= 1:
                allc[i][4] += -1;
        queue += -1;
    
    # update the simulation time and the service time for each iteration.
    simtime += tinterval;
    simtime = round(simtime, 2);
    if tellerstate == True:
        servicecount += tinterval;
        servicecount = round(servicecount, 2);
    queuem += queue;

# calculations for customer queue times and time spent at the bank.
queuelen = queuem/401;
queuett = 0; 
banktt = 0;
for i in range (0,10):
    allc[i][7] = round((allc[i][5]-allc[i][1]),2);
    allc[i][8] = round((allc[i][6]-allc[i][1]),2);
    del allc[i][3];
    del allc[i][3];
    print('\n',allc[i],'\n');
    queuett += allc[i][5];
    banktt += allc[i][6];
queueav = round(queuett/10,4);
bankav = round(banktt/10,4);

print('\n Total teller service time = ',tellertime,'\n');
print('\n Average time spent in queue = ',queueav,'\n');
print('\n Average length queue = ',queuelen,'\n');
print('\n Average time spent in bank = ',bankav,'\n');
    
# entry 1: Customer ID
# entry 2: Arrival time
# entry 3: Length of service
# entry 4: Time of commencement of service
# entry 5: Departure time
# entry 6: Time in queue
# entry 7: Time in bank
```
